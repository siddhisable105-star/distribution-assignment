
1. Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use
BINOM.DIST)

-    1. Probability of rolling a 3 exactly 5 times (Excel simulation + BINOM.DIST)
Number of trials (n) = 30
Probability of success (rolling 3) (p) = 1/6
Required: P(X = 5)

In Excel:=BINOM.DIST(5, 30, 1/6, FALSE)
This gives the probability of getting exactly 5 times “3” in 30 rolls.

 Simulation step:
Use =RANDBETWEEN(1,6) in 30 cells → count 3’s → repeat multiple times.

3. A dataset has a mean of 50 and a standard deviation of 5. What percentage of values lie between 45 and
55 if the data follows a normal distribution?

-    Percentage between 45 and 55 (Normal Distribution)

Given:
Mean (μ) = 50
SD (σ) = 5

Use Z-score:

Z₁ = (45−50)/5 = −1
Z₂ = (55−50)/5 = +1

 From normal distribution:

Area between −1 and +1 = 68% (approx)

✔ Answer: ~68% of values

4. What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset

-   Standardization (Z-score)

 Formula:
z = \frac{x - \mu}{\sigma}


Meaning:
Converts data into standard units
Mean becomes 0, SD becomes 1
Importance:
Compare different datasets
Detect outliers
Used in machine learning & statistics


5. What is Kurtosis and their type?

-  Kurtosis and its types

 Kurtosis measures peakedness or tail heaviness of data.

Types:
Mesokurtic → Normal shape (kurtosis ≈ 3)
Leptokurtic → High peak, heavy tails
Platykurtic → Flat peak, light tails

6. Explain why the uniform distribution is a good model for the outcome of rolling a fair die
-  Why uniform distribution fits a fair die?
A fair die has 6 outcomes: 1–6

Each has equal probability = 1/6

 This matches uniform distribution (equal likelihood)

 7. Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5
 -  Probability of at least 8 successes (Excel)

Given:

n = 15
p = 0.5

Excel formula:=1 - BINOM.DIST(7, 15, 0.5, TRUE)

This gives P(X ≥ 8)

8. How does log transformation help in stabilizing variance and making data more normally distributed?

-  Log transformation (why useful?)

 Purpose:

Reduces skewness
Stabilizes variance
Makes data closer to normal distribution

Example:
Before: Highly right-skewed data

Benefits:
